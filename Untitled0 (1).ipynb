```python
def is_float(x):
  try:
    float(x)
  except:
    return False
  return True
```

```python
def covert(x):
 tokens = x.split('-')
 if len(tokens) == 2:
    return (float(tokens[0])+float(tokens[1]))/2
 try:
    return float(x)
 except:
    return None
```

```python
df4= df3.copy()
df4['total_sqft']=df4['total_sqft'].apply(covert)
df4
```

```python
df4['price per sqft']=df4['price']*100000/df4['total_sqft']
df4
```

```python

def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf['price per sqft'])
        s = np.std(subdf['price per sqft'])
        reduced_df = subdf[(subdf['price per sqft'] > (m - s)) & (subdf['price per sqft'] <= (m + s))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out


df6 = remove_pps_outliers(df4)
print(df6.shape)


def plot_scatter_chart(df, location):
    bhk2 = df[(df.location == location) & (df.bhk == 2)]
    bhk3 = df[(df.location == location) & (df.bhk == 3)]
    matplotlib.rcParams['figure.figsize'] = (15, 10)
    plt.scatter(bhk2.total_sqft, bhk2.price, color='blue', label='2 BHK', s=50)
    plt.scatter(bhk3.total_sqft, bhk3.price, color='red', label='3 BHK', s=50)
    plt.xlabel('total_sqft')
    plt.ylabel('price')
    plt.title(location)
    plt.legend()
    plt.show()


plot_scatter_chart(df6, 'Rajaji Nagar')

```

```python


df7 = df6.drop(['size', 'price per sqft'], axis='columns')
df7.head()



dummies= pd.get_dummies(df7.location)
dummies.head()



df8= pd.concat([df7, dummies.drop('other',axis='columns')],axis='columns')



dff8=df8.drop('location',axis='columns')
df8.head()


X = df8.drop(['price', 'location', 'area_type', 'society'], axis='columns')



'
availability_dummies = pd.get_dummies(X['availability'], prefix='availability')


X = X.drop('availability', axis='columns')
X = pd.concat([X, availability_dummies], axis='columns')

X.head()

Y = df8.price
Y.head()

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=10)

lr_clf = LinearRegression()
lr_clf.fit(X_train, Y_train)
lr_clf.score(X_test, Y_test)
```

```python


from sklearn.ensemble import RandomForestRegressor


rf_clf = RandomForestRegressor(n_estimators=5000, random_state=10)


rf_clf.fit(X_train, Y_train)


rf_score = rf_clf.score(X_test, Y_test)
print(f"Random Forest Regressor Score: {rf_score}")
```

```python
def predict_price(location, sqft, bath, bhk):
    loc_index = np.where(X.columns == location)[0][0]
    x=np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1
    return lr_clf.predict([x])[0]
```
