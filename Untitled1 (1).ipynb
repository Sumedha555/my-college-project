```python
import streamlit as st
import re
import pandas as pd
import zipfile
import os
import tempfile # To handle file uploads temporarily
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import nltk
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from transformers import pipeline # For transformer sentiment (optional)
import plotly.express as px # Using Plotly for interactive plots (optional)


@st.cache_resource # Cache this resource to avoid re-downloading
def download_nltk_data():
    try:
        nltk.data.find('corpora/stopwords')
    except nltk.downloader.DownloadError:
        nltk.download('stopwords')
    try:
        nltk.data.find('sentiment/vader_lexicon.zip')
    except nltk.downloader.DownloadError:
         nltk.download('vader_lexicon')
    st.success("NLTK resources downloaded.")

download_nltk_data()

# Initialize Sentiment Analyzers
vader = SentimentIntensityAnalyzer()

# Initialize a Transformer-based sentiment analyzer (e.g., 'sentiment-analysis')
# This can be computationally intensive. Consider adding a toggle in the app
# or providing a choice of sentiment analysis methods.
@st.cache_resource # Cache the transformer pipeline
def load_transformer_model():
    try:
        # Reduce precision to 'fp16' if you have a suitable GPU and need less memory
        model = pipeline('sentiment-analysis', device=-1) # device=-1 for CPU, device=0 for GPU
        st.success("Transformer sentiment analyzer loaded (using CPU/GPU if available).")
        return model
    except Exception as e:
        st.warning(f"Could not load Transformer sentiment analyzer: {e}. VADER will be used for sentiment.")
        return None

# transformer_sentiment_analyzer = load_transformer_model() # Load only if needed

def parse_whatsapp_chat(chat_file_path):
    """Parses a WhatsApp chat text file into a pandas DataFrame."""
    # Adjust this pattern based on your chat file format!
    # Common patterns are listed in the previous response.
    pattern = re.compile(r'^\[?(\d{1,2}[-/]\d{1,2}[-/]\d{2,4}),?\s*(\d{1,2}:\d{2}(?::\d{2})?(?: [AP]M)?)\]?\s*(.*?): (.*)$')

    data = []
    unmatched_lines = []

    with open(chat_file_path, 'r', encoding='utf-8') as f:
        for line in f:
            match = pattern.match(line)
            if match:
                try:
                    date, time, sender, message = match.groups()
                    data.append([date, time, sender, message])
                except ValueError:
                    unmatched_lines.append(line.strip())
            else:
                unmatched_lines.append(line.strip())

    df = pd.DataFrame(data, columns=['Date', 'Time', 'Sender', 'Message'])

    try:
        df['Timestamp'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], errors='coerce', infer_datetime_format=True)
        st.write(f"Successfully parsed {df['Timestamp'].notna().sum()} timestamps out of {len(df)} rows.")
    except Exception as e:
        st.error(f"Error converting to datetime: {e}. Timestamp column may be incomplete.")

    return df, unmatched_lines

def analyze_chat(df):
    """Performs the analysis on the DataFrame."""
    if df.empty:
        st.warning("DataFrame is empty. No analysis to perform.")
        return

    st.header("Overall Statistics")
    total_messages = len(df)
    st.write(f"Total number of messages: {total_messages}")

    st.header("Media Messages")
    media_messages = df[df['Message'] == '<Media omitted>']
    total_media_count = len(media_messages)
    st.write(f"Total media messages (sent or received): {total_media_count}")
    if total_media_count > 0:
        media_by_sender = media_messages.groupby('Sender').size().reset_index(name='Media Count')
        st.write("Media messages sent/received per sender:")
        st.dataframe(media_by_sender)
    else:
        st.write("No media messages found.")

    st.header("Word Count")
    # Calculate the total number of words
    df['word_count'] = df['Message'].astype(str).apply(lambda x: len(x.split())) # Handle potential non-string messages

    total_words = df['word_count'].sum()
    st.write(f"Total number of words in the chat: {total_words}")

    st.header("Most Common Words (Word Cloud)")
    if not df['Message'].empty:
        all_messages = ' '.join(df['Message'].astype(str).dropna()) # Handle NaN messages

        # Remove unwanted patterns like '<Media omitted>'
        all_messages = re.sub(r'<Media omitted>', '', all_messages)
        all_messages = re.sub(r'http\S+', '', all_messages) # Remove URLs

        # Tokenize words and remove stopwords
        words = re.findall(r'\b\w+\b', all_messages.lower())
        stop_words = set(stopwords.words('english'))
        filtered_words = [word for word in words if word not in stop_words]

        if filtered_words:
            word_counts = Counter(filtered_words)
            most_common_words = word_counts.most_common(10)
            st.write("Most common words:")
            st.write(most_common_words)

            # Generate and display the word cloud
            wordcloud = WordCloud(width = 800, height = 800,
                                  background_color ='white',
                                  stopwords = stop_words,
                                  min_font_size = 10).generate_from_frequencies(word_counts)

            plt.figure(figsize = (8, 8), facecolor = None)
            plt.imshow(wordcloud)
            plt.axis("off")
            plt.tight_layout(pad = 0)
            st.pyplot(plt)
        else:
            st.write("No significant words found after filtering.")
    else:
        st.write("No messages available for word analysis.")

    st.header("Links Exchanged")
    if not df['Message'].empty:
        url_pattern = re.compile(r'https?://\S+|www\.\S+')
        df['Link_Count'] = df['Message'].astype(str).apply(lambda x: len(url_pattern.findall(x)))
        links_exchanged_df = df[df['Link_Count'] > 0].copy()
        total_links_exchanged = links_exchanged_df['Link_Count'].sum()

        st.write(f"Total number of links exchanged: {total_links_exchanged}")

        if not links_exchanged_df.empty:
            links_by_sender = links_exchanged_df.groupby('Sender')['Link_Count'].sum().reset_index(name='Total Links Sent')
            st.write("Total links sent by each sender:")
            st.dataframe(links_by_sender)
        else:
            st.write("No messages containing links were found.")
    else:
         st.write("No messages available for link analysis.")


    st.header("Timeline Analysis")
    if 'Timestamp' in df.columns and pd.api.types.is_datetime64_any_dtype(df['Timestamp']) and df['Timestamp'].notna().sum() > 0:
        messages_by_year = df['Timestamp'].dt.year.value_counts().sort_index()
        st.write("Messages by Year:")
        st.write(messages_by_year)

        messages_over_time = df.set_index('Timestamp').resample('M').size()
        st.write("Messages over Time (Monthly):")
        # Use Plotly for better interactivity in Streamlit
        fig = px.line(messages_over_time, x=messages_over_time.index, y=messages_over_time.values, title='Total Messages Over Time (Monthly)')
        fig.update_layout(xaxis_title="Date", yaxis_title="Number of Messages")
        st.plotly_chart(fig)

    else:
        st.warning("Cannot analyze timeline. 'Timestamp' column is missing, invalid, or contains no valid dates.")


    st.header("Sentiment Analysis (VADER)")

    # Apply VADER sentiment analysis
    st.write("Applying VADER sentiment analysis...")
    df['vader_sentiment_score'] = df['Message'].astype(str).apply(lambda x: vader.polarity_scores(x)['compound'])
    st.write("VADER sentiment analysis complete.")


    if 'vader_sentiment_score' in df.columns and 'Timestamp' in df.columns and df['Timestamp'].notna().sum() > 0:
        daily_sentiment = df.set_index('Timestamp')['vader_sentiment_score'].resample('D').mean().reset_index()
        daily_sentiment.columns = ['Date', 'Average_VADER_Sentiment']

        st.write("Average VADER Sentiment per day:")
        st.dataframe(daily_sentiment.head())

        fig = px.line(daily_sentiment, x='Date', y='Average_VADER_Sentiment', title='Daily Average VADER Sentiment Trend')
        fig.update_layout(xaxis_title="Date", yaxis_title="Average Sentiment Score (VADER)")
        st.plotly_chart(fig)

        st.subheader("Sentiment Distribution")
        # Function to categorize sentiment
        def categorize_sentiment(score):
            if score >= 0.05:
                return 'Happy'
            elif score <= -0.05:
                return 'Angry'
            else:
                return 'Sad'

        df['vader_sentiment_category'] = df['vader_sentiment_score'].apply(categorize_sentiment)
        sentiment_counts = df['vader_sentiment_category'].value_counts()

        st.write("Sentiment Distribution (based on VADER compound score):")
        st.write(sentiment_counts)

        if not sentiment_counts.empty:
            fig = px.pie(values=sentiment_counts.values, names=sentiment_counts.index, title='Distribution of Sentiment (Happy, Sad, Angry)')
            st.plotly_chart(fig)
        else:
            st.write("No sentiment data to plot.")


    else:
        st.warning("Cannot perform VADER sentiment analysis. Ensure 'Message' and 'Timestamp' columns are available and 'Timestamp' has valid dates.")


    # Optional: Transformer Sentiment Analysis (Can be slow)
    # if st.checkbox("Run Transformer Sentiment Analysis (may be slow)"):
    #     transformer_sentiment_analyzer = load_transformer_model()
    #     if transformer_sentiment_analyzer:
    #         st.write("Applying Transformer sentiment analysis...")
    #         # Apply in batches or sample if dataset is very large
    #         try:
    #             # Limiting text length for transformer
    #             df['transformer_sentiment_score'] = df['Message'].astype(str).apply(lambda x: get_transformer_sentiment(x[:512]))
    #             st.write("Transformer sentiment analysis complete.")
    #             # You would then analyze and display results similar to VADER
    #         except Exception as e:
    #             st.error(f"An error occurred during Transformer sentiment application: {e}")
    # else:
    #     st.info("Transformer sentiment analysis skipped.")


# Main Streamlit App
st.title("WhatsApp Chat Analyzer")

uploaded_file = st.file_uploader("Upload your WhatsApp chat zip file", type="zip")

if uploaded_file is not None:
    st.success("File uploaded successfully!")

    # Use tempfile to handle the uploaded zip file
    with tempfile.TemporaryDirectory() as tmpdir:
        extract_dir = os.path.join(tmpdir, "whatsapp_chat")
        os.makedirs(extract_dir, exist_ok=True)

        try:
            # Save the uploaded file to the temporary directory
            zip_path = os.path.join(tmpdir, uploaded_file.name)
            with open(zip_path, "wb") as f:
                f.write(uploaded_file.getbuffer())

            # Unzip the file
            with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                zip_ref.extractall(extract_dir)
            st.write(f"Successfully extracted chat to {extract_dir}")

            # Find the chat file
            chat_file = None
            for root, _, files in os.walk(extract_dir):
                for file in files:
                    # Adjust 'WhatsApp Chat' if your file name is different.
                    if file.endswith('.txt') and 'WhatsApp Chat' in file:
                        chat_file = os.path.join(root, file)
                        break
                if chat_file:
                    break

            if chat_file:
                st.write(f"Found chat file: {chat_file}")
                st.write("Parsing chat file...")

                # Parse the chat
                df, unmatched_lines = parse_whatsapp_chat(chat_file)

                if unmatched_lines:
                     st.warning(f"{len(unmatched_lines)} lines did not match the regex pattern. Analysis might be incomplete.")
                     with st.expander("Show first 10 unmatched lines"):
                         for i in range(min(10, len(unmatched_lines))):
                             st.text(unmatched_lines[i])
                     st.info("If the DataFrame is empty or has few rows, inspect the unmatched lines to refine the regex pattern in the code.")


                # Perform analysis
                analyze_chat(df)


            else:
                st.error("Error: Could not find a .txt chat file in the extracted directory.")

        except FileNotFoundError:
            st.error(f"Error: File not found after upload.")
        except zipfile.BadZipFile:
            st.error(f"Error: Bad zip file. Could not open the uploaded file.")
        except Exception as e:
            st.error(f"An unexpected error occurred: {e}")
```

```python
!pip install streamlit plotly

import streamlit as st
import re
import pandas as pd
import zipfile
import os
import tempfile # To handle file uploads temporarily
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import nltk
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from transformers import pipeline # For transformer sentiment (optional)
import plotly.express as px # Using Plotly for interactive plots (optional)

# Ensure NLTK resources are downloaded
@st.cache_resource # Cache this resource to avoid re-downloading
def download_nltk_data():
    try:
        nltk.data.find('corpora/stopwords')
    except nltk.downloader.DownloadError:
        nltk.download('stopwords')
    try:
        nltk.data.find('sentiment/vader_lexicon.zip')
    except nltk.downloader.DownloadError:
         nltk.download('vader_lexicon')
    st.success("NLTK resources downloaded.")

download_nltk_data()

# Initialize Sentiment Analyzers
vader = SentimentIntensityAnalyzer()

# Initialize a Transformer-based sentiment analyzer (e.g., 'sentiment-analysis')
# This can be computationally intensive. Consider adding a toggle in the app
# or providing a choice of sentiment analysis methods.
@st.cache_resource # Cache the transformer pipeline
def load_transformer_model():
    try:
        # Reduce precision to 'fp16' if you have a suitable GPU and need less memory
        model = pipeline('sentiment-analysis', device=-1) # device=-1 for CPU, device=0 for GPU
        st.success("Transformer sentiment analyzer loaded (using CPU/GPU if available).")
        return model
    except Exception as e:
        st.warning(f"Could not load Transformer sentiment analyzer: {e}. VADER will be used for sentiment.")
        return None

# transformer_sentiment_analyzer = load_transformer_model() # Load only if needed

def parse_whatsapp_chat(chat_file_path):
    """Parses a WhatsApp chat text file into a pandas DataFrame."""
    # Adjust this pattern based on your chat file format!
    # Common patterns are listed in the previous response.
    pattern = re.compile(r'^\[?(\d{1,2}[-/]\d{1,2}[-/]\d{2,4}),?\s*(\d{1,2}:\d{2}(?::\d{2})?(?: [AP]M)?)\]?\s*(.*?): (.*)$')

    data = []
    unmatched_lines = []

    with open(chat_file_path, 'r', encoding='utf-8') as f:
        for line in f:
            match = pattern.match(line)
            if match:
                try:
                    date, time, sender, message = match.groups()
                    data.append([date, time, sender, message])
                except ValueError:
                    unmatched_lines.append(line.strip())
            else:
                unmatched_lines.append(line.strip())

    df = pd.DataFrame(data, columns=['Date', 'Time', 'Sender', 'Message'])

    try:
        df['Timestamp'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], errors='coerce', infer_datetime_format=True)
        st.write(f"Successfully parsed {df['Timestamp'].notna().sum()} timestamps out of {len(df)} rows.")
    except Exception as e:
        st.error(f"Error converting to datetime: {e}. Timestamp column may be incomplete.")

    return df, unmatched_lines

def analyze_chat(df):
    """Performs the analysis on the DataFrame."""
    if df.empty:
        st.warning("DataFrame is empty. No analysis to perform.")
        return

    st.header("Overall Statistics")
    total_messages = len(df)
    st.write(f"Total number of messages: {total_messages}")

    st.header("Media Messages")
    media_messages = df[df['Message'] == '<Media omitted>']
    total_media_count = len(media_messages)
    st.write(f"Total media messages (sent or received): {total_media_count}")
    if total_media_count > 0:
        media_by_sender = media_messages.groupby('Sender').size().reset_index(name='Media Count')
        st.write("Media messages sent/received per sender:")
        st.dataframe(media_by_sender)
    else:
        st.write("No media messages found.")

    st.header("Word Count")
    # Calculate the total number of words
    df['word_count'] = df['Message'].astype(str).apply(lambda x: len(x.split())) # Handle potential non-string messages

    total_words = df['word_count'].sum()
    st.write(f"Total number of words in the chat: {total_words}")

    st.header("Most Common Words (Word Cloud)")
    if not df['Message'].empty:
        all_messages = ' '.join(df['Message'].astype(str).dropna()) # Handle NaN messages

        # Remove unwanted patterns like '<Media omitted>'
        all_messages = re.sub(r'<Media omitted>', '', all_messages)
        all_messages = re.sub(r'http\S+', '', all_messages) # Remove URLs

        # Tokenize words and remove stopwords
        words = re.findall(r'\b\w+\b', all_messages.lower())
        stop_words = set(stopwords.words('english'))
        filtered_words = [word for word in words if word not in stop_words]

        if filtered_words:
            word_counts = Counter(filtered_words)
            most_common_words = word_counts.most_common(10)
            st.write("Most common words:")
            st.write(most_common_words)

            # Generate and display the word cloud
            wordcloud = WordCloud(width = 800, height = 800,
                                  background_color ='white',
                                  stopwords = stop_words,
                                  min_font_size = 10).generate_from_frequencies(word_counts)

            plt.figure(figsize = (8, 8), facecolor = None)
            plt.imshow(wordcloud)
            plt.axis("off")
            plt.tight_layout(pad = 0)
            st.pyplot(plt)
        else:
            st.write("No significant words found after filtering.")
    else:
        st.write("No messages available for word analysis.")

    st.header("Links Exchanged")
    if not df['Message'].empty:
        url_pattern = re.compile(r'https?://\S+|www\.\S+')
        df['Link_Count'] = df['Message'].astype(str).apply(lambda x: len(url_pattern.findall(x)))
        links_exchanged_df = df[df['Link_Count'] > 0].copy()
        total_links_exchanged = links_exchanged_df['Link_Count'].sum()

        st.write(f"Total number of links exchanged: {total_links_exchanged}")

        if not links_exchanged_df.empty:
            links_by_sender = links_exchanged_df.groupby('Sender')['Link_Count'].sum().reset_index(name='Total Links Sent')
            st.write("Total links sent by each sender:")
            st.dataframe(links_by_sender)
        else:
            st.write("No messages containing links were found.")
    else:
         st.write("No messages available for link analysis.")


    st.header("Timeline Analysis")
    if 'Timestamp' in df.columns and pd.api.types.is_datetime64_any_dtype(df['Timestamp']) and df['Timestamp'].notna().sum() > 0:
        messages_by_year = df['Timestamp'].dt.year.value_counts().sort_index()
        st.write("Messages by Year:")
        st.write(messages_by_year)

        messages_over_time = df.set_index('Timestamp').resample('M').size()
        st.write("Messages over Time (Monthly):")
        # Use Plotly for better interactivity in Streamlit
        fig = px.line(messages_over_time, x=messages_over_time.index, y=messages_over_time.values, title='Total Messages Over Time (Monthly)')
        fig.update_layout(xaxis_title="Date", yaxis_title="Number of Messages")
        st.plotly_chart(fig)

    else:
        st.warning("Cannot analyze timeline. 'Timestamp' column is missing, invalid, or contains no valid dates.")


    st.header("Sentiment Analysis (VADER)")

    # Apply VADER sentiment analysis
    st.write("Applying VADER sentiment analysis...")
    df['vader_sentiment_score'] = df['Message'].astype(str).apply(lambda x: vader.polarity_scores(x)['compound'])
    st.write("VADER sentiment analysis complete.")


    if 'vader_sentiment_score' in df.columns and 'Timestamp' in df.columns and df['Timestamp'].notna().sum() > 0:
        daily_sentiment = df.set_index('Timestamp')['vader_sentiment_score'].resample('D').mean().reset_index()
        daily_sentiment.columns = ['Date', 'Average_VADER_Sentiment']

        st.write("Average VADER Sentiment per day:")
        st.dataframe(daily_sentiment.head())

        fig = px.line(daily_sentiment, x='Date', y='Average_VADER_Sentiment', title='Daily Average VADER Sentiment Trend')
        fig.update_layout(xaxis_title="Date", yaxis_title="Average Sentiment Score (VADER)")
        st.plotly_chart(fig)

        st.subheader("Sentiment Distribution")
        # Function to categorize sentiment
        def categorize_sentiment(score):
            if score >= 0.05:
                return 'Happy'
            elif score <= -0.05:
                return 'Angry'
            else:
                return 'Sad'

        df['vader_sentiment_category'] = df['vader_sentiment_score'].apply(categorize_sentiment)
        sentiment_counts = df['vader_sentiment_category'].value_counts()

        st.write("Sentiment Distribution (based on VADER compound score):")
        st.write(sentiment_counts)

        if not sentiment_counts.empty:
            fig = px.pie(values=sentiment_counts.values, names=sentiment_counts.index, title='Distribution of Sentiment (Happy, Sad, Angry)')
            st.plotly_chart(fig)
        else:
            st.write("No sentiment data to plot.")


    else:
        st.warning("Cannot perform VADER sentiment analysis. Ensure 'Message' and 'Timestamp' columns are available and 'Timestamp' has valid dates.")





# Main Streamlit App
st.title("WhatsApp Chat Analyzer")

uploaded_file = st.file_uploader("Upload your WhatsApp chat zip file", type="zip")

if uploaded_file is not None:
    st.success("File uploaded successfully!")

    # Use tempfile to handle the uploaded zip file
    with tempfile.TemporaryDirectory() as tmpdir:
        extract_dir = os.path.join(tmpdir, "whatsapp_chat")
        os.makedirs(extract_dir, exist_ok=True)

        try:
            # Save the uploaded file to the temporary directory
            zip_path = os.path.join(tmpdir, uploaded_file.name)
            with open(zip_path, "wb") as f:
                f.write(uploaded_file.getbuffer())

            # Unzip the file
            with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                zip_ref.extractall(extract_dir)
            st.write(f"Successfully extracted chat to {extract_dir}")

            # Find the chat file
            chat_file = None
            for root, _, files in os.walk(extract_dir):
                for file in files:
                    # Adjust 'WhatsApp Chat' if your file name is different.
                    if file.endswith('.txt') and 'WhatsApp Chat' in file:
                        chat_file = os.path.join(root, file)
                        break
                if chat_file:
                    break

            if chat_file:
                st.write(f"Found chat file: {chat_file}")
                st.write("Parsing chat file...")

                # Parse the chat
                df, unmatched_lines = parse_whatsapp_chat(chat_file)

                if unmatched_lines:
                     st.warning(f"{len(unmatched_lines)} lines did not match the regex pattern. Analysis might be incomplete.")
                     with st.expander("Show first 10 unmatched lines"):
                         for i in range(min(10, len(unmatched_lines))):
                             st.text(unmatched_lines[i])
                     st.info("If the DataFrame is empty or has few rows, inspect the unmatched lines to refine the regex pattern in the code.")


                # Perform analysis
                analyze_chat(df)


            else:
                st.error("Error: Could not find a .txt chat file in the extracted directory.")

        except FileNotFoundError:
            st.error(f"Error: File not found after upload.")
        except zipfile.BadZipFile:
            st.error(f"Error: Bad zip file. Could not open the uploaded file.")
        except Exception as e:
            st.error(f"An unexpected error occurred: {e}")
```
